```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
keypoints = ["Nose", "Ear", "Shoulder", "Elbow", "Wrist", "Pinky", "Index", "Thumb", "Hip", "Knee", "Ankle", "Heel", "Toes"]
less_noisy = {"Nose": 93.2, "Ear": 94.8, "Shoulder": 93.2, "Elbow": 93.7, "Wrist": 93.8, "Pinky": 86.9, "Index\nfinger": 90.7, "Thumb": 88.4, "Hip": 96.4, "Knee": 95.1, "Ankle": 95.6, "Heel": 94.4, "Toes": 94.8 }
more_noisy = {"Nose": 91.8, "Ear": 93.0, "Shoulder": 93.0, "Elbow": 93.8, "Wrist": 85.5, "Pinky": 85.0, "Index\nfinger": 89.6, "Thumb": 86.4, "Hip": 94.9, "Knee": 94.4, "Ankle": 95.2, "Heel": 93.4, "Toes": 93.3}

ind = np.arange(len(keypoints))
width = 0.35

fig, ax  = plt.subplots()
rects1 = ax.bar(ind - width/2, less_noisy.values(), width, label="Less noisy")
rects2 = ax.bar(ind + width/2, more_noisy.values(), width, label="More noisy")

ax.set_ylabel("PCK@0.1")
ax.set_title("Mean performance of keypoints")
ax.set_xticks(ind)
ax.set_xticklabels(keypoints)
ax.legend(bbox_to_anchor=(1, 1))

#fig.tight_layout()
plt.xticks(rotation=45)

plt.savefig("./entities/kpts.pdf", bbox_inches = "tight")
plt.show()
```
